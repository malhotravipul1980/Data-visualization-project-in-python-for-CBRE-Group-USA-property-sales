```python

Q1=np.quantile(df["Total_Basement_Area"],0.25)
Q3=np.quantile(df["Total_Basement_Area"],0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
#print(lower_bound,upper_bound)
df["Total_Basement_Area"]=np.where(df["Total_Basement_Area"]<lower_bound,lower_bound,df["Total_Basement_Area"])
df["Total_Basement_Area"]=np.where(df["Total_Basement_Area"]>upper_bound,upper_bound,df["Total_Basement_Area"])
```

```python
sns.boxplot(y=df["Total_Basement_Area"])
```

```python
for i in contVar:
    Q1=np.quantile(df[i],0.25)
    Q3=np.quantile(df[i],0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    df[i]=np.where(df[i]<lower_bound,lower_bound,df[i])
    df[i]=np.where(df[i]>upper_bound,upper_bound,df[i])
```

```python
for i in contVar:
    plt.figure()
    sns.boxplot(y=df[i])
```
